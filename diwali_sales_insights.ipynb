```python
# Correlation Matrix
numeric_columns = ['Age', 'Orders', 'Amount']
correlation_matrix = df[numeric_columns].corr()
print("Correlation Analysis:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix: Age vs Orders vs Spending')
plt.show()
```

Interpretation: All correlations are very weak (< 0.03), indicating that age and order frequency alone don't predict spending behavior. This tells us we need to look at other factors like occupation, location, and product categories for meaningful insights.



```python
# Customer Segmentation based on Spending
print("\nCUSTOMER VALUE SEGMENTATION:")
df['Customer_Value'] = pd.cut(df['Amount'],
                             bins=[0, 5000, 10000, 20000, df['Amount'].max()],
                             labels=['Low', 'Medium', 'High', 'Premium'])

value_segment = df['Customer_Value'].value_counts()
print("Customer Value Distribution:")
for segment, count in value_segment.items():
    percentage = (count / len(df)) * 100
    print(f"• {segment} Value: {count} customers ({percentage:.1f}%)")

# Visualization
plt.figure(figsize=(10, 6))
sns.countplot(x='Customer_Value', data=df, order=['Low', 'Medium', 'High', 'Premium'])
plt.title('Customer Value Segmentation')
plt.xlabel('Customer Value Tier')
plt.ylabel('Number of Customers')
for bars in plt.gca().containers:
    plt.gca().bar_label(bars)
plt.show()
```

Interpretation: We successfully segmented customers into 4 value tiers. The Medium Value segment (43.5%) is largest, while Premium customers (3.7%) represent the most valuable group. This segmentation helps in targeted marketing strategies.


```python
# Age Group Statistical Analysis
print("\nAGE GROUP STATISTICS:")
age_stats = df.groupby('Age Group')['Amount'].agg(['mean', 'median', 'std', 'count']).round(2)
print("Spending Patterns by Age Group:")
print(age_stats)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Customer Count by Age Group
age_count = df['Age Group'].value_counts()
sns.barplot(x=age_count.index, y=age_count.values, ax=ax1)
ax1.set_title('Customer Distribution by Age Group')
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Number of Customers')
for bars in ax1.containers:
    ax1.bar_label(bars)

# Spending by Age Group
age_spending = df.groupby('Age Group', as_index=False)['Amount'].sum()
sns.barplot(x='Age Group', y='Amount', data=age_spending, ax=ax2)
ax2.set_title('Total Spending by Age Group')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Total Amount (₹)')

plt.tight_layout()
plt.show()
```

Interpretation: Our age analysis reveals a strategic dual opportunity: the 26-35 segment (4,541 customers) provides the volume foundation of our business, while the 51-55 segment, though smaller, delivers superior average spending power (₹9,954 vs ₹9,384). This suggests implementing tiered marketing strategies—volume-focused campaigns for the younger segment and premium, high-value offerings for the mature segment to maximize overall revenue potential.

```python
# Gender-wise Performance
print("\nGENDER PERFORMANCE ANALYSIS:")
gender_stats = df.groupby('Gender').agg({
    'Amount': ['mean', 'sum', 'count'],
    'Orders': 'mean'
}).round(2)
gender_stats.columns = ['Avg_Spending', 'Total_Revenue', 'Customer_Count', 'Avg_Orders']
print(gender_stats)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Customer Count by Gender
sns.countplot(x='Gender', data=df, ax=ax1)
ax1.set_title('Customer Count by Gender')
for bars in ax1.containers:
    ax1.bar_label(bars)

# Total Spending by Gender
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum()
sns.barplot(x='Gender', y='Amount', data=sales_gen, ax=ax2)
ax2.set_title('Total Spending by Gender')
plt.tight_layout()
plt.show()
```

Interpretation: Our data reveals a clear pattern: women are driving the majority of Diwali sales. Nearly 7 out of every 10 customers are women, and they contribute to 70% of our total revenue. This isn't just a trend—it's the core of our customer base that we should nurture and understand better. They also spend slightly more on average (₹9,491 vs ₹9,367). This clear gender preference should guide marketing focus.



```python
# Top Performing States
print("\nTOP STATES PERFORMANCE:")
state_performance = df.groupby('State').agg(
    Total_Orders=('Orders', 'sum'),
    Total_Revenue=('Amount', 'sum'),
    Avg_Order_Value=('Amount', 'mean'),
    Total_Transactions=('User_ID', 'count')
).reset_index()
top_states_table = state_performance.nlargest(5, 'Total_Revenue')

print("Top 5 Performing States by Revenue:")
print(top_states_table[['State', 'Total_Revenue', 'Total_Orders']])
top_states_revenue = df.groupby('State')['Amount'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_states_revenue.values, y=top_states_revenue.index)
# ... code to label the plot ...
plt.show()
```

Interpretation: Uttar Pradesh, Maharashtra, and Karnataka are the top 3 revenue-generating states, contributing significantly to overall sales. Delhi shows the highest average order value (₹10,512), indicating premium purchasing behavior.



```python
# Occupation-wise Analysis
print("\nOCCUPATION ANALYSIS:")
occupation_stats = df.groupby('Occupation').agg({
    'Amount': ['sum', 'mean', 'count'],
    'Orders': 'sum'
}).round(2)
occupation_stats.columns = ['Total_Revenue', 'Avg_Spending', 'Customer_Count', 'Total_Orders']
occupation_stats = occupation_stats.sort_values('Total_Revenue', ascending=False)
print("Top Occupations by Revenue:")
print(occupation_stats.head(6))

# Visualization
plt.figure(figsize=(12, 6))
top_occ = occupation_stats.head(6).reset_index()
sns.barplot(x='Total_Revenue', y='Occupation', data=top_occ)
plt.title('Top 6 Occupations by Revenue')
plt.xlabel('Total Revenue (₹)')
plt.tight_layout()
plt.show()
```

Interpretation: IT Sector, Healthcare, and Aviation professionals are the top 3 spending occupations. Despite Govt employees having fewer customers, they show the highest average spending (₹9,973), making them valuable for premium targeting.



```python
# Product Category Analysis
print("\nPRODUCT CATEGORY PERFORMANCE:")
category_metrics = df.groupby('Product_Category').agg({
    'Orders': 'sum',
    'Amount': ['sum', 'mean', 'count']
}).round(2)
category_metrics.columns = ['Total_Orders', 'Total_Revenue', 'Avg_Order_Value', 'Transaction_Count']
top_categories = category_metrics.nlargest(5, 'Total_Revenue')
print("Top 5 Product Categories by Revenue:")
print(top_categories)

# Visualization
plt.figure(figsize=(12, 6))
top_cat_viz = df.groupby('Product_Category')['Amount'].sum().nlargest(10).reset_index()
sns.barplot(x='Amount', y='Product_Category', data=top_cat_viz)
plt.title('Top 10 Product Categories by Revenue')
plt.xlabel('Total Revenue (₹)')
plt.tight_layout()
plt.show()
```

Interpretation: Food category generates the highest revenue (₹33.9M) with substantial orders. Furniture has the highest average order value (₹15,455) despite fewer transactions, indicating high-value purchases in this category.
